```python
import matplotlib.pyplot as plt
```

```python
apl_price=[93.95,112.15,104.05,144.85,164.69]
ms_price=[39.01,50.29,57.05,69.68,94.39]
year=[2014,2015,2016,2017,2018]
```

```python
plt.plot(year,apl_price)
plt.plot(year,ms_price)
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```python
plt.scatter(year,apl_price)
plt.scatter(year,ms_price)
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```python
plt.plot(year,apl_price,
         year, ms_price)
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```python
# k is for black color and g for green -- for dashed
# o is for scatter
plt.plot(year,apl_price,'--k'
        ,year,ms_price,'og')
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```python
#define limits with plt.axis

plt.plot(year,apl_price,'--k'
        ,year,ms_price,'og')
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.axis ([2013,2019,20,200])
plt.show()
```

```python
fig_1= plt.figure(1,figsize=(20,9.6))

chart_1=fig_1.add_subplot(221)
chart_2=fig_1.add_subplot(224)

chart_1.plot(year,apl_price,'k')
chart_2.scatter(year,ms_price)
plt.show()
```

# STACK PLOT

```python
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

```

```python
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','cyan','red','black'])
```

```python
plt.plot([],[],color='m',label='sleeping')
plt.plot([],[],color='c',label='eating')
plt.plot([],[],color='r',label='working')
plt.plot([],[],color='k',label='playing')
plt.legend()
```

```python
#combine both to get legends
plt.plot([],[],color='m',label='sleeping',linewidth=10)
plt.plot([],[],color='c',label='eating',linewidth=10)
plt.plot([],[],color='r',label='working',linewidth=10)
plt.plot([],[],color='k',label='playing',linewidth=10)
plt.legend()
plt.xlabel('Day')
plt.ylabel('Number of hours Spent')
plt.title('Check it out')
plt.xticks(days)
plt.yticks(np.arange(25))
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','cyan','red','black'])

plt.show()
```

```python
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=90)
plt.title('Day last Activities')
plt.show()
```

```python
#draw eating slice. use explode

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
       explode=(0,0,0,0.1),
       autopct='%.1f%%')
plt.title('Day last Activities')
plt.show()
```
